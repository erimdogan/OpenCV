```python
import cv2
import numpy as np
```

# Basic Operations

```python
img = cv2.imread("Resources/lena.png")
```

## Convert Image to Gray

```python
imgGRY = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("Gray Image", imgGRY) # Gri tonlamalı resim
cv2.waitKey(0)
```

## Blur

```python
imgBlur = cv2.GaussianBlur(imgGRY, (7,7), 0)
cv2.imshow("Blur Image", imgBlur) # Blur Image
cv2.waitKey(0)
```

## Canny

```python
imgCanny = cv2.Canny(img, 100, 100)
cv2.imshow("Canny Image", imgCanny) # Canny kenar algılama
cv2.waitKey(0)
```

## Dialotion

```python
kernel = np.ones((5,5), np.uint8)
imgDialotion = cv2.dilate(imgCanny, kernel, iterations=1) # resimdeki beyaz kısımları genişletir
cv2.imshow("Dialotion Image", imgDialotion)
cv2.waitKey(0)
```

## Erotion

```python
imgEroded =cv2.erode(imgDialotion, kernel, iterations=1) # resimdeki beyaz kısımları daraltır
cv2.imshow("Eroded Image", imgEroded)
cv2.waitKey(0)
```

x axes is similar to normal x axes, but positive y axes is going down differ from normal y axes
